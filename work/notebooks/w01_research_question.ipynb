## 2. The question: decision, action, cost of a wrong call

Question: Which observable signals are most strongly associated with page decline?

Decision: Prioritize which pages content teams should review first for potential refresh.

Action: Content team investigates top-ranked pages, checks why they're declining, decides refresh/rewrite/merge/monitor.

Cost of wrong call:
- False positive (recommending a page that's fine): wastes reviewer time
- False negative (missing a declining page): lost traffic/revenue continues

```python
import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
df['is_declining'] = (df['trend_direction'] == 'down').astype(int)

print(f"Declining pages: {df['is_declining'].mean()*100:.1f}%")

for col in ['impressions_90d', 'ctr', 'avg_position', 'content_age_days']:
    print(f"{col}: {df[col].corr(df['is_declining']):.3f}")
```
